```python
# 1.데이터 준비 및 전처리 
#라이브러리 불러오기
import time #학습 시간을 측정하기 위해 
from konlpy.tag import Okt #한국어 형태소 분석기
from gensim.models import Word2Vec
from sklearn.decomposition import PCA # 차원 축소
import matplotlib.pyplot as plt #식각화 
import re #정규 표현식 
```

```python
# 2.데이터 로드 및 전처리 
#fairy_tale.txt 파일을 읽어온 후 전처리 및 토큰화 
with open('fairy_tale.txt', 'r', encoding='utf-8') as f:
    text = f.read()
print('test=', text)
```

```python
text = text.replace('## 배경 이야기','')
print(text)
```

```python
#소문자 변환 (사실 한국어이므로 큰 영향 없음)
text = text.lower()

#특수문자 제거 
text = re.sub(r'[^가-힣\s]','',text)
#[^...] 이것빼고 다 지워! 이런 뜻 
#즉, [^가-힣\s] 한글이랑 공백빼고 다 지워 ! 

#Okt 형태소 분석기 초기화 
okt = Okt()

# 문장 단위로 분리 후 형태소 단위 토큰화 
sentences = text.split('\n')
print('text after split=', sentences)

```

```python
del sentences[0]
print('text after spllit=', sentences)
```

```python
from collections import Counter

tokens = okt.morphs(text) # 형태소 단위로 자름

# 단어 빈도 세기
word_counts = Counter(tokens)

# 가장 많이 나온 단어 30개 출력
print("💬 가장 많이 등장한 단어:")
for word, count in word_counts.most_common(10):
    print(f"{word} : {count}회")
```

```python
# stop_words = "을 의 는 이 가 를 그"
stop_words = "이 로 어느 에 의 가 와"
```

```python
tokenized = []
for sent in sentences:
    #형태소 분석
    tokens = okt.morphs(sent, stem=True) 
    #stem = True 단어들을 기본형으로 바꿔준다. 

    #길이 1 이하 토큰 제거 -> 하면 안된다. 용, 성 등 중요 키워드 삭제된다. 
    # tokens = [t for t in tokens if len(t) > 1]
    # print('tokens after okt & len <= 1 remove =', tokens)

    #불용어 제거 (split 안하면 문자열이 하나로 이어져서 인식되기 때문에 split(" ")해줌
    tokens = [word for word in tokens if word not in stop_words.split(" ")]
    print("tokens after stop_words", tokens)

    tokenized.append(tokens)

print('첫 문장 토큰 예시:', tokenized[0])
```

## CBOW(Continuous Bag of Words)
- 주변이 있는 단어들을 입력으로 중간에 있는 단어들을 예측하는 방법

## Skip-Gram
- 중간에 있는 단어들을 입력으로 주변 단어들을 예측하는 방법 


```python
#CBOW 모델 학습 
start_time = time.time()
model_cbow = Word2Vec(sentences=tokenized,
                      vector_size=50,
                      window=3,
                      min_count=2,
                      sg=0,
                      workers=4,
                      seed=42)
cbow_time = time.time() - start_time
print(f'CBOW 학습 시간: {cbow_time:.2f}초')
```

```python
#skip-gram 모델 학습 
start_time = time.time()
model_sg = Word2Vec(sentences=tokenized,
                    vector_size=50,
                    window=3,
                    min_count=2,
                    sg=1,
                    workers=4,
                    seed =42)
sg_time = time.time() - start_time
print(f'Skip-gram 학습시간: {sg_time:.2f}초')
```

```python

# 5. 단어 유사도 출력
keywords = ['왕자', '공주', '마법', '용', '성']

print('CBOW 유사 단어:')
for word in keywords:
    try:
        sims = model_cbow.wv.most_similar(word, topn=5)
        print(f'{word}: {sims}')
    except KeyError:
        print(f'{word}: 코퍼스에 없음')

print('\nSkip-gram 유사 단어:')
for word in keywords:
    try:
        sims = model_sg.wv.most_similar(word, topn=5)
        print(f'{word}: {sims}')
    except KeyError:
        print(f'{word}: 코퍼스에 없음')
```

```python
print('용' in [t for sublist in tokenized for t in sublist])
print('성' in [t for sublist in tokenized for t in sublist])
```

```python
# 주피터 노트북 한글 깨짐 문제 해결: 주피터 노트북(Jupyter Notebook)
import platform
from matplotlib import font_manager, rc

path = "c:/Windows/Fonts/malgun.ttf"
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')

# 한글 깨짐 문제 해결: 공통. minus 처리.
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
```

```python

# 6. 임베딩 시각화 (PCA)
# 각 단어 벡터 가져오기
words = keywords
vectors = [model_sg.wv[w] for w in words if w in model_sg.wv]

# PCA로 2차원 축소
pca = PCA(n_components=2)
coords = pca.fit_transform(vectors)

# 플롯
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    if word in model_sg.wv:
        x, y = coords[i]
        plt.scatter(x, y)
        plt.text(x+0.01, y+0.01, word, fontsize=12)
plt.title('Skip-gram 임베딩 PCA 시각화')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()
```

```python
# 6. 임베딩 시각화 (PCA)
# 각 단어 벡터 가져오기
words = keywords
vectors = [model_cbow.wv[w] for w in words if w in model_cbow.wv]

# PCA로 2차원 축소
pca = PCA(n_components=2)
coords = pca.fit_transform(vectors)

# 플롯
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    if word in model_cbow.wv:
        x, y = coords[i]
        plt.scatter(x, y)
        plt.text(x+0.01, y+0.01, word, fontsize=12)
plt.title('cbow 임베딩 PCA 시각화')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()
```

```python
# CBOW 유사도
cbow_sim = model_cbow.wv.similarity('왕자', '공주')
print(f"CBOW - '왕자' vs '공주' 유사도: {cbow_sim:.4f}")

# Skip-gram 유사도
sg_sim = model_sg.wv.similarity('왕자', '공주')
print(f"Skip-gram - '왕자' vs '공주' 유사도: {sg_sim:.4f}")
```

✅ 동화 코퍼스엔 어느 쪽이 적합할까?
- 동화 텍스트처럼 문장이 짧고 반복되는 경우, CBOW가 더 적합하다.
- CBOW는 자주 등장하는 단어 조합을 기반으로 빠르게 안정적인 벡터를 학습함
- Skip-gram은 희귀 단어에 민감하지만, 동화같이 단순하고 반복 많은 코퍼스에서는 오히려 덜 효과적일 수 있음


## cbow 와 skip-gram 유사도 히트맵

```python
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

words = ['왕자', '공주', '마법', '용', '성', '탑']
vectors = [model_cbow.wv[word] for word in words]
sim_matrix = cosine_similarity(vectors)

sns.heatmap(sim_matrix, xticklabels=words, yticklabels=words, annot=True, cmap='YlGnBu')
plt.title("CBOW 유사도 히트맵")
plt.show()
```
