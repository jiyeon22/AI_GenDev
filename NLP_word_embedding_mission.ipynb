```python
# 1.ë°ì´í„° ì¤€ë¹„ ë° ì „ì²˜ë¦¬ 
#ë¼ì´ë¸ŒëŸ¬ë¦¬ ë¶ˆëŸ¬ì˜¤ê¸°
import time #í•™ìŠµ ì‹œê°„ì„ ì¸¡ì •í•˜ê¸° ìœ„í•´ 
from konlpy.tag import Okt #í•œêµ­ì–´ í˜•íƒœì†Œ ë¶„ì„ê¸°
from gensim.models import Word2Vec
from sklearn.decomposition import PCA # ì°¨ì› ì¶•ì†Œ
import matplotlib.pyplot as plt #ì‹ê°í™” 
import re #ì •ê·œ í‘œí˜„ì‹ 
```

```python
# 2.ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬ 
#fairy_tale.txt íŒŒì¼ì„ ì½ì–´ì˜¨ í›„ ì „ì²˜ë¦¬ ë° í† í°í™” 
with open('fairy_tale.txt', 'r', encoding='utf-8') as f:
    text = f.read()
print('test=', text)
```

```python
text = text.replace('## ë°°ê²½ ì´ì•¼ê¸°','')
print(text)
```

```python
#ì†Œë¬¸ìž ë³€í™˜ (ì‚¬ì‹¤ í•œêµ­ì–´ì´ë¯€ë¡œ í° ì˜í–¥ ì—†ìŒ)
text = text.lower()

#íŠ¹ìˆ˜ë¬¸ìž ì œê±° 
text = re.sub(r'[^ê°€-íž£\s]','',text)
#[^...] ì´ê²ƒë¹¼ê³  ë‹¤ ì§€ì›Œ! ì´ëŸ° ëœ» 
#ì¦‰, [^ê°€-íž£\s] í•œê¸€ì´ëž‘ ê³µë°±ë¹¼ê³  ë‹¤ ì§€ì›Œ ! 

#Okt í˜•íƒœì†Œ ë¶„ì„ê¸° ì´ˆê¸°í™” 
okt = Okt()

# ë¬¸ìž¥ ë‹¨ìœ„ë¡œ ë¶„ë¦¬ í›„ í˜•íƒœì†Œ ë‹¨ìœ„ í† í°í™” 
sentences = text.split('\n')
print('text after split=', sentences)

```

```python
del sentences[0]
print('text after spllit=', sentences)
```

```python
from collections import Counter

tokens = okt.morphs(text) # í˜•íƒœì†Œ ë‹¨ìœ„ë¡œ ìžë¦„

# ë‹¨ì–´ ë¹ˆë„ ì„¸ê¸°
word_counts = Counter(tokens)

# ê°€ìž¥ ë§Žì´ ë‚˜ì˜¨ ë‹¨ì–´ 30ê°œ ì¶œë ¥
print("ðŸ’¬ ê°€ìž¥ ë§Žì´ ë“±ìž¥í•œ ë‹¨ì–´:")
for word, count in word_counts.most_common(10):
    print(f"{word} : {count}íšŒ")
```

```python
# stop_words = "ì„ ì˜ ëŠ” ì´ ê°€ ë¥¼ ê·¸"
stop_words = "ì´ ë¡œ ì–´ëŠ ì— ì˜ ê°€ ì™€"
```

```python
tokenized = []
for sent in sentences:
    #í˜•íƒœì†Œ ë¶„ì„
    tokens = okt.morphs(sent, stem=True) 
    #stem = True ë‹¨ì–´ë“¤ì„ ê¸°ë³¸í˜•ìœ¼ë¡œ ë°”ê¿”ì¤€ë‹¤. 

    #ê¸¸ì´ 1 ì´í•˜ í† í° ì œê±° -> í•˜ë©´ ì•ˆëœë‹¤. ìš©, ì„± ë“± ì¤‘ìš” í‚¤ì›Œë“œ ì‚­ì œëœë‹¤. 
    # tokens = [t for t in tokens if len(t) > 1]
    # print('tokens after okt & len <= 1 remove =', tokens)

    #ë¶ˆìš©ì–´ ì œê±° (split ì•ˆí•˜ë©´ ë¬¸ìžì—´ì´ í•˜ë‚˜ë¡œ ì´ì–´ì ¸ì„œ ì¸ì‹ë˜ê¸° ë•Œë¬¸ì— split(" ")í•´ì¤Œ
    tokens = [word for word in tokens if word not in stop_words.split(" ")]
    print("tokens after stop_words", tokens)

    tokenized.append(tokens)

print('ì²« ë¬¸ìž¥ í† í° ì˜ˆì‹œ:', tokenized[0])
```

## CBOW(Continuous Bag of Words)
- ì£¼ë³€ì´ ìžˆëŠ” ë‹¨ì–´ë“¤ì„ ìž…ë ¥ìœ¼ë¡œ ì¤‘ê°„ì— ìžˆëŠ” ë‹¨ì–´ë“¤ì„ ì˜ˆì¸¡í•˜ëŠ” ë°©ë²•

## Skip-Gram
- ì¤‘ê°„ì— ìžˆëŠ” ë‹¨ì–´ë“¤ì„ ìž…ë ¥ìœ¼ë¡œ ì£¼ë³€ ë‹¨ì–´ë“¤ì„ ì˜ˆì¸¡í•˜ëŠ” ë°©ë²• 


```python
#CBOW ëª¨ë¸ í•™ìŠµ 
start_time = time.time()
model_cbow = Word2Vec(sentences=tokenized,
                      vector_size=50,
                      window=3,
                      min_count=2,
                      sg=0,
                      workers=4,
                      seed=42)
cbow_time = time.time() - start_time
print(f'CBOW í•™ìŠµ ì‹œê°„: {cbow_time:.2f}ì´ˆ')
```

```python
#skip-gram ëª¨ë¸ í•™ìŠµ 
start_time = time.time()
model_sg = Word2Vec(sentences=tokenized,
                    vector_size=50,
                    window=3,
                    min_count=2,
                    sg=1,
                    workers=4,
                    seed =42)
sg_time = time.time() - start_time
print(f'Skip-gram í•™ìŠµì‹œê°„: {sg_time:.2f}ì´ˆ')
```

```python

# 5. ë‹¨ì–´ ìœ ì‚¬ë„ ì¶œë ¥
keywords = ['ì™•ìž', 'ê³µì£¼', 'ë§ˆë²•', 'ìš©', 'ì„±']

print('CBOW ìœ ì‚¬ ë‹¨ì–´:')
for word in keywords:
    try:
        sims = model_cbow.wv.most_similar(word, topn=5)
        print(f'{word}: {sims}')
    except KeyError:
        print(f'{word}: ì½”í¼ìŠ¤ì— ì—†ìŒ')

print('\nSkip-gram ìœ ì‚¬ ë‹¨ì–´:')
for word in keywords:
    try:
        sims = model_sg.wv.most_similar(word, topn=5)
        print(f'{word}: {sims}')
    except KeyError:
        print(f'{word}: ì½”í¼ìŠ¤ì— ì—†ìŒ')
```

```python
# ì£¼í”¼í„° ë…¸íŠ¸ë¶ í•œê¸€ ê¹¨ì§ ë¬¸ì œ í•´ê²°: ì£¼í”¼í„° ë…¸íŠ¸ë¶(Jupyter Notebook)
import platform
from matplotlib import font_manager, rc

path = "c:/Windows/Fonts/malgun.ttf"
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')

# í•œê¸€ ê¹¨ì§ ë¬¸ì œ í•´ê²°: ê³µí†µ. minus ì²˜ë¦¬.
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus'] = False
```

```python

# 6. ìž„ë² ë”© ì‹œê°í™” (PCA)
# ê° ë‹¨ì–´ ë²¡í„° ê°€ì ¸ì˜¤ê¸°
words = keywords
vectors = [model_sg.wv[w] for w in words if w in model_sg.wv]

# PCAë¡œ 2ì°¨ì› ì¶•ì†Œ
pca = PCA(n_components=2)
coords = pca.fit_transform(vectors)

# í”Œë¡¯
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    if word in model_sg.wv:
        x, y = coords[i]
        plt.scatter(x, y)
        plt.text(x+0.01, y+0.01, word, fontsize=12)
plt.title('Skip-gram ìž„ë² ë”© PCA ì‹œê°í™”')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()
```

```python
# 6. ìž„ë² ë”© ì‹œê°í™” (PCA)
# ê° ë‹¨ì–´ ë²¡í„° ê°€ì ¸ì˜¤ê¸°
words = keywords
vectors = [model_cbow.wv[w] for w in words if w in model_cbow.wv]

# PCAë¡œ 2ì°¨ì› ì¶•ì†Œ
pca = PCA(n_components=2)
coords = pca.fit_transform(vectors)

# í”Œë¡¯
plt.figure(figsize=(8, 6))
for i, word in enumerate(words):
    if word in model_cbow.wv:
        x, y = coords[i]
        plt.scatter(x, y)
        plt.text(x+0.01, y+0.01, word, fontsize=12)
plt.title('cbow ìž„ë² ë”© PCA ì‹œê°í™”')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.grid(True)
plt.show()
```

```python
# CBOW ìœ ì‚¬ë„
cbow_sim = model_cbow.wv.similarity('ì™•ìž', 'ê³µì£¼')
print(f"CBOW - 'ì™•ìž' vs 'ê³µì£¼' ìœ ì‚¬ë„: {cbow_sim:.4f}")

# Skip-gram ìœ ì‚¬ë„
sg_sim = model_sg.wv.similarity('ì™•ìž', 'ê³µì£¼')
print(f"Skip-gram - 'ì™•ìž' vs 'ê³µì£¼' ìœ ì‚¬ë„: {sg_sim:.4f}")
```
